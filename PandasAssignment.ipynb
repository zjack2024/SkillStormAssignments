```python
import pandas as pd
```

```python
df = pd.read_csv('20240617-DE-TS-LectureMaterials/Python/Data/imdb_top_1000.csv')
```

```python
nf =df.drop(["Poster_Link","Certificate","No_of_Votes","Meta_score",'Runtime'], axis="columns")
nf
```

```python
year = nf[pd.to_numeric(nf['Released_Year'],errors='coerce').notnull()]
oldest = year.sort_values(by=['Released_Year'])
oldest.head(1)
```

```python
newest = year.sort_values(by=['Released_Year'],ascending=False)
newest.head(1)
```

```python
rate=nf.sort_values(by=['IMDB_Rating'],ascending=False)
rate.head(10)
```

```python
nf2=nf.copy()
nf2['Genres']= nf2['Genre'].str.split(', ').tolist()
nf2=nf2.explode('Genres')
genres= nf2.sort_values(by=['IMDB_Rating'],ascending=False)
genreTop=genres.drop_duplicates(subset=['Genres'], keep='first')
genreBest=genreTop.drop_duplicates(subset=['Series_Title'], keep='first')
genreBest.head(20)
```

```python
director=nf.value_counts('Director')
director.head(1)
```

```python
actors1=nf.copy()
actors1['Stars']=actors1[['Star1','Star2','Star3','Star4']].values.tolist()
actors2=actors1.explode('Stars')
stars=actors2.value_counts('Stars')
stars.head(1)
```

```python
nf3=nf2.copy()
nf3=nf3.dropna(subset=['Gross'])
nf3['Gross'] = nf3.Gross.str.split(',').str.join('').astype(int)
nf3.head()
```

```python
nf4=nf3.copy()
nf4=nf3.dropna(subset=['Gross'])
genresGross= nf4.sort_values(by=['Gross'],ascending=False)
genresGross1=genresGross.drop_duplicates(subset=['Genres'], keep='first')
genresGross2=genresGross1.drop_duplicates(subset=['Series_Title'], keep='first')
genresGross2.head(25)
```

```python
directorGross=nf4.sort_values(by=['Gross'])
directors=directorGross.drop_duplicates(subset=['Director'],keep='first')
directors.head(10)
```
